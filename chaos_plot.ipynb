### [1] 
Plot of function : 
$x_{n+1} = a x_n(1-x_n)$

where $a = [0.5, 1.5, 3.3, 4] $ and $x_0 = 0.35$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#function to plot :
f = lambda x,a: a*x*(1-x)
a = np.array([0.5, 1.5, 3.3, 4])
n = 100
x = np.zeros(n+1)
x[0] = 0.35
```

```python
#fill the value of x[i] for i=1...100
for i in range(1, len(x)):
    x[i] = f(x[i-1], a[0])
#plot
plt.rcParams['figure.figsize'] = [13, 10]
n_array = np.array([i for i in range(n+1)])
plt.title("a="+str(a[0]))
plt.xlabel('n values (0-100)')
plt.ylabel('x values')
plt.plot(n_array, x, '.-', color='red')
plt.show()
```

### [2]
Plot of $x_i$ against a, where :

$i = 0,...,50$, $x_0 = 0.35$, and $0\leq a \leq 4$, $\delta a = 0.01$

```python
#plot of Sa vs a, where 0<=a<=4, delta(a)=0.01, n = 50
a_array = np.linspace(0, 4, num=401)
n = 50
x_array = np.zeros((n+1, len(a_array)))

#fill initial values for each a as x = 0.35
for i in range(x_array.shape[1]):
    x_array[0][i] = 0.35
x_array = x_array.T

for i in range(x_array.shape[0]):
    for j in range(1, x_array.shape[1]):
        x_array[i][j] = f(x_array[i][j-1], a_array[i])
        
#reduce the array
a_array = np.repeat(a_array, x_array.shape[1])
x_array = x_array.flatten()

#plot the array
plt.rcParams.update({'font.size': 22})
plt.rcParams['figure.figsize'] = [25, 20]
plt.xlabel('a values (0<=a<=4)')
plt.ylabel('xi values, with x0 = 0.35, i=0,...,50')
plt.title('xi vs a')
plt.plot(a_array, x_array, '.', color='red')
plt.show()
```
